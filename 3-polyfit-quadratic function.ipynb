```python
import numpy as np
import matplotlib.pyplot as plt
#!pip install scikit-learn
```

```python
#setup data
x_coord = range(-10,10)
y_coord =[x*x for x in x_coord]
plt.scatter(x_coord,y_coord)

#plt.plot(x_coord,y_coord, "o")
plt.show()
```

```python
#fit 
my_model = np.poly1d(np.polyfit(x_coord,y_coord,2))
#trend
my_line = np.linspace(-10,10,100)
```

```python
#polynomial regression
plt.scatter(x_coord,y_coord)
plt.plot(my_line, my_model(my_line))
```

```python
#Estimate X, Y relationship using R-square
#r2 = [0,1]. 0: no relationship, 1:100% relationship
```

```python
import sklearn
from sklearn.metrics import r2_score
print(r2_score(y_coord, my_model(x_coord)))
```

```python
#Predict future values
#set any value in the my_model and see what happens
next_val = my_model(11)
print (next_val)
```
