# Create a training set of 10 (x,y) pairs

```python
#allows a tuple of values as input
f2 = np.vectorize(y_generator)

#pick 100 random numbers
np.random.seed(100)

#pick 100 x values for x_training dataset
x_train = np.sort(np.random.uniform(-10,1,100))

#generate corresponding y
y_train = f2(x_train)
```

```python
#plot training data
plt.plot(x_train,y_train,'o')
plt.xlabel('X Training Data')
plt.ylabel('Y Training Data')
plt.title('2D plot for Training Data')
plt.show()
```

# Test dataset with 100 pairs

```python
x_test = np.random.uniform(-10,1,100)
y_test = f2(x_test)
#plot data
plt.plot(x_test,y_test, 'X')
plt.xlabel('X Testing Data')
plt.ylabel('Y Testing Data')
plt.title ('2D plot with testing dataset')
plt.show()
```

```python
# Fit the data with polynomial of degree 1, 2, 3
```

```python
#degree 1 
x_new = np.linspace(-15,1,15)
y_new = f2(x_new)
my_model = np.poly1d(np.polyfit(x_new,y_new,1))
plt.scatter(x_train, y_train)
plt.plot(x_new, my_model(x_new), 'r', label="Fit "+str(1)+ "degree poly")
plt.show()

#goodness of the fit
print (mean_squared_error(y_new, my_model(x_new), squared=False))
```

```python
#degree 2
x_new = np.linspace(-15,1,15)
y_new = f2(x_new)
my_model2 = np.poly1d(np.polyfit(x_new, y_new, 2))
plt.scatter(x_train, y_train)
plt.plot(x_new, my_model2(x_new), 'r', label="Fit "+str(2)+ "degree poly")
plt.show()

#goodness of the fit
print (mean_squared_error(y_new, my_model2(x_new), squared=False))
```

```python
#degree 3
x_new = np.linspace(-15,1,15)
y_new = f2(x_new)
my_model3 = np.poly1d(np.polyfit(x_new,y_new,3))
plt.scatter(x_train, y_train)
plt.plot(x_new, my_model3(x_new), 'r', label="Fit "+str(3)+ "degree poly")
plt.show()
```

```python
#degree 9
x_new = np.linspace(-15,1,15)
y_new = f2(x_new)
my_model9 = np.poly1d(np.polyfit(x_new,y_new,9))
plt.scatter(x_train, y_train)
plt.plot(x_new, my_model9(x_new), 'r', label="Fit "+str(9)+ "degree poly")
plt.show()

#goodness of the fit
print (mean_squared_error(y_new, my_model9(x_new), squared=False))
```
