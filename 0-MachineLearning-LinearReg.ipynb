# Linear Regression 

```python
import numpy as np
import matplotlib.pyplot as plt
#setup dataset and model using polyfit 

x = [89, 43, 28, 71, 52, 50, 62, 71, 15, 22, 27, 36]
y = [121, 43, 12, 17, 25, 15 ,26, 18, 51, 22, 27, 63]
```

```python
#plot the data (first step)
#plt.scatter(x,y)
```

```python
#create a data model
data_model = np.poly1d(np.polyfit(x,y,1))
#generate the polynomial that fits the dataset 
print(data_model)
```

```python
#To test the model using some new data, create 100 new data points in the interval
new_data = np.linspace(min(x),max(x),100)
```

```python
#now check how the new_data fits the data_model
plt.scatter(x,y)
plt.plot(new_data, data_model(new_data))
plt.show()
```

```python
#calculate RMSE
import sklearn 
from sklearn.metrics import r2_score # higher the score better correlation
print (r2_score(y, data_model(x)))
```

```python
#Let's try n=2
data_model_2 = np.poly1d(np.polyfit(x,y,2))
plt.scatter(x,y)
plt.plot(new_data, data_model_2(new_data))
plt.show()
print (r2_score(y,data_model_2(x)))

#n=3
data_model_3 = np.poly1d(np.polyfit(x,y,3))
plt.scatter(x,y)
plt.plot(new_data, data_model_3(new_data))
plt.show()
print (r2_score(y,data_model_3(x)))
```

```python
#Let's calculate RMSE 
from sklearn.metrics import mean_squared_error #lower the score better is the prediction
#n=1
print (mean_squared_error(y, data_model(x), squared=False))
#n=2
print (mean_squared_error(y, data_model_2(x), squared=False))
#n=5
data_model_5 = np.poly1d(np.polyfit(x,y,5))
print (mean_squared_error(y, data_model_5(x), squared=False))
```
